```python
import pandas as pd
import numpy as np

df = pd.read_csv("loadTxt/advance_kyogi_load.txt")

docs1=np.array([])
for x in df.values:
  docs1=np.append(docs1,x)  
```

```python
docs=np.array([])
for x in df.values:
  if (any(map(str.isdigit, x[0]))):
    docs=np.append(docs,x)
```

```python
vectorizer = TfidfVectorizer(use_idf=True,token_pattern=u'(?u)\\b\\w+\\b')
vecs = vectorizer.fit_transform(docs1)
TF_list=vecs.toarray()

words = vectorizer.get_feature_names()

doc_list=[]
for doc_id, vec in zip(range(len(docs)), vecs.toarray()):
    docs=[]
    count=0
    for w_id, tfidf in sorted(enumerate(vec), key=lambda x: x[1], reverse=True):
        lemma = words[w_id]
        if tfidf > 0.00:
          doc=[lemma,tfidf]
          docs.append(doc)
          count=count+1
    doc=[doc_id,count]
    docs.insert(0,doc)
    doc_list.append(docs)
```

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("../../../data/Colab Notebooks/testdata/testdata010/kyogi_rule.txt")

docsKNP=np.array([])
for x in df1.values:
  docsKNP=np.append(docsKNP,x)
#KNPテキストをarrayに挿入
```

```python
listw=[]
for x in docsKNP:
  word_list=[]
  word=''
  for y in x:
    if y!=' ':
      word +=y
    else :
      word_list.append(word)
      word=''
  word_list.append(word)
  listw.append(word_list)
#' 'でスプリットし、KNP形式のリストを生成
```

```python
alist=[]
slist=[]
plist=[]
wlist=bp=ba=bs=''
list_=[]
sword=['#','*','+']
for x in listw:
  if x!=[]:  
    if not x[0] in sword and not x[0]=='EOS':
        wlist+=x[0]

    if x[0] in sword:
      if wlist != '':
        list_.append(bp)
        list_.append(wlist)
        wlist=''
        plist.append(list_)
        list_=[]
      bp=x[1]

    if x[0] == '*' or x[0]=='#':
      if plist != []:
        list_.append(ba)
        list_.append(plist)
        plist=[]
        slist.append(list_)
        list_=[]
      ba=x[1]

    if x[0] == '#': 
      if slist != []:
        list_.append(bs)
        list_.append(slist)
        slist=[]
        alist.append(list_)
        list_=[]
      bs=x[1]
#KNP形式リストを構造体で管理
```

```python
pointList=[]
apointList=[]
apoint=[]
point=[]

phase=''
for x in alist:
  for y in x[1]:#*
    for z in y[1]:#+
      point.append(z[1])
      phase+=z[1]
    apoint.append(phase)
    phase=''
  apointList.append(apoint) 
  pointList.append(point)
  apoint=[]
  point=[]
#*と+ごとに語句リストを生成
```

```python
docs2=np.array([])
for x in pointList:
  for y in x:
    if (any(map(str.isdigit, y))):
      docs2=np.append(docs2,y)
#数字を含む単語を抽出
```

```python
addresList=[]
addres=[]
scount=0
wcount=0

for x in docs2:
  scount=0
  for y in pointList:
    wcount=0
    for z in y:
      if x==z:
        addres.append(scount)
        addres.append(wcount)
        addres.append(z)
        addresList.append(addres)
        addres=[]
      wcount+=1
    scount+=1
#数字を含む単語(+単位)の文中の番地を抽出
addresList.sort()
```

```python
def get_unique_list(seq):
    seen = []
    return [x for x in seq if x not in seen and not seen.append(x)]
```

```python
addresList.sort()
addresList=get_unique_list(addresList)
#重複の削除およびソート

number=[]
numberList=[]
ncount=1
while ncount>=1:
  count=0
  ncount=0
  for x in addresList:
    number=addresList.pop(count)
    numberList.append(number)
    count+=1
    ncount+=1

numberList=get_unique_list(numberList)
numberList.sort()
```

```python
def get_list(seq,search_word):
    if type(seq[1]) is list:
      return get_list(seq[1],search_word) 
    elif seq[1]==search_word:
      return seq[0]
    elif seq[1]!=search_word:
      return get_list(seq,search_word)
    return seq,search_word
```

```python
def id_of_in(seq,word):
  for ph in seq[1]:
    for pair in  ph[1]:
      if pair[1]==word:
        return pair[0]
```

```python
def words_of_id(seq,id_word):
  list_=[]
  for ph in seq[1]:
    for pair in ph[1]:
      if id_word in pair[0] and len(id_word)+1==len(pair[0]):
        list_.append(pair[1])
  return list_
```
