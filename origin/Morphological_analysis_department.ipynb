```
import pandas as pd
import numpy as np

df = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/load.txt")

docs1=np.array([])
for x in df.values:
  docs1=np.append(docs1,x)  

docs1
```

```
docs_num=np.array([])
for x in df.values:
  if (any(map(str.isdigit, x[0]))):
    docs_num=np.append(docs_num,x)

#数字が含まれる文章の抽出
docs_num
```

```
vectorizer = TfidfVectorizer(use_idf=True,token_pattern=u'(?u)\\b\\w+\\b')
vecs = vectorizer.fit_transform(docs1)
TF_list=vecs.toarray()

words = vectorizer.get_feature_names()

doc_list=[]
for doc_id, vec in zip(range(len(docs1)), vecs.toarray()):
    docs=[]
    count=0
    for w_id, tfidf in sorted(enumerate(vec), key=lambda x: x[1], reverse=True):
        lemma = words[w_id]
        if tfidf > 0.00:
          doc=[lemma,tfidf]
          docs.append(doc)
          count=count+1
    doc=[doc_id,count]
    docs.insert(0,doc)
    doc_list.append(docs)
```

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# vectorizeする
vectorizer.fit(docs1)

#抽出した単語を確認する
wlist1=vectorizer.get_feature_names()
# ['a', 'am', 'and', 'are', 'i', 'man', 'perfect', 'regend', robot', 'sam', 'you']
#単語のリスト化

countlist = [0] * len(wlist1)
```

```
i=0
lists=[]
for x in TF_list:
  for y in x:
    if y>0.0:
      list_=[]
      list_.append(wlist1[i])
      list_.append(y)
      countlist[i]=countlist[i]+1
      lists.append(list_)
    i=i+1
  i=0
      #単語とTF-IDF値の追加
lists.sort()
lists
```

```
#出現回数の追加

list_=[]
for x in lists:
  for j,y in enumerate(wlist1):
    if x[0] == y:
      x.append(countlist[j])

lists
```

```
import re

#優先値の追加
for x in lists:
  weight=0
  # ひらがなの抽出
  hiragana = re.findall("[ぁ-ん]", x[0])
  # カタカナの抽出
  katakana = re.findall("[ァ-ン]", x[0])
  # 漢字の抽出
  kanji = re.findall("[一-龥]", x[0])

  haifun = re.findall("[ー]", x[0])

  eigo = re.findall("[a-z]", x[0])

  suuzi = re.findall("[a-z]", x[0])

  weight=len(x[0])+len(hiragana)*0.1+(len(katakana)+len(haifun)+len(eigo)+len(suuzi))*0.2+len(kanji)

  x.append(weight)

lists
```

```
for x in lists:
  count=0
  for y in lists:
    for z in y:
      if type(z) is str:
        if x[0] in z and x[0] != z:
          count=count+1
  x.append(count)

lists
```

```
#判定結果を追加

for x in lists:
  x.insert(1,"0")
```

```
#抽出リスト完成
lists
```

```
#出力
new_list = sorted(lists,reverse=True)

list_=["単語","判定結果","TF-IDF値","出現回数","優先値","連結回数"]
new_list.insert(0,list_)

with open("/content/gdrive/My Drive/Colab Notebooks/extracted_list.csv", 'w') as f:
    writer = csv.writer(f, lineterminator='\n')
    for x in new_list:
        writer.writerow(x)
#機械学習のデータセット出力
```
